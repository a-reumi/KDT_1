[ 국민 해외 관광객 성별 ] <hr>

```python
## [1-1] 모듈 로딩
import pandas as pd       # 데이터 분석 및 전처리 관련 모듈
import numpy as np        # 배열(array) 파이썬 구현 관련 모듈 
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
## [1-2] 데이터 준비
DATA_FILE1 = './data/20250121182446_국민 해외관광객 성별1.csv' 
```

```python
## 데이터 로딩
gtDF1=pd.read_csv(DATA_FILE1) 
gtDF1=gtDF1.drop(gtDF1[gtDF1['성별']=='승무원'].index) 
#gtDF1.loc[gtDF1['성별'].isin(['전체']),['성별','기준일자','인원','증감률'] ] 
gtDF1.loc[:,['기준일자','성별','인원','증감률'] ] 
```

```python
yData1 = []
yData2 = []
gtDF1=gtDF1.drop(gtDF1[gtDF1['기준일자']=='전체'].index)
gtDF1=gtDF1.drop(gtDF1[gtDF1['성별']=='전체'].index)
xData = gtDF1['기준일자'].unique() 

gtDF1.reset_index(drop=True, inplace=True) 
for idx in gtDF1.index:
    if idx %2 == 0: 
        yData1.append(gtDF1.loc[idx, '인원'])
    else:
        yData2.append(gtDF1.loc[idx, '인원']) 

plt.plot(xData, yData1, label='남성', color='#ff71ce')
plt.plot(xData, yData2, label='여성', color='#01cdfe')
plt.title('국민 해외관광객 성별')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.legend()

```

[ 국민 해외 관광객 연령별 ] <hr>

```python
## [1-2] 데이터 준비
DATA_FILE2 = './data/20250121184834_국민 해외관광객 연령별2.csv'
```

```python
## 데이터 로딩
gtDF2=pd.read_csv(DATA_FILE2) 
gtDF2.head(10)
gt2=gtDF2.drop(gtDF2[gtDF2['연령']=='승무원'].index) 
# gtDF2.loc[gtDF2['연령'].isin(['전체']),['기준일자','연령','인원','증감률'] ] 
gt2.loc[:,['기준일자','연령','인원','증감률'] ] 
```

```python

gtDF2=gtDF2.drop(gtDF2[gtDF2['기준일자']=='전체'].index)
gtDF2=gtDF2.drop(gtDF2[gtDF2['연령']=='전체'].index) 
gtDF2=gtDF2.drop(gtDF2[gtDF2['연령']=='승무원'].index) 
gt2=gtDF2.groupby(['연령'])['인원'].sum()
gt2=pd.DataFrame(gt2)


plt.barh(gt2.index, gt2.loc[gt2.index, '인원'], color='#ffeead') 
plt.ticklabel_format(axis='x',useOffset=False, style='plain')
plt.title('국민 해외관광객 연령별')
```

```python
## 실제 데이터
print(gtDF2.head(10), end='\n\n')
print(gtDF2.tail(10))
```

[국민 해외 관광객 교통수단별] <hr> 

```python
## [1-2] 데이터 준비
DATA_FILE3 = './data/20250121183142_국민 해외관광객 교통수단별3.csv'
```

```python
## 데이터 로딩
gtDF3=pd.read_csv(DATA_FILE3) 
gtDF3.head()
gtDF3.loc[:,['기준일자','교통수단구분','인원','증감률'] ] 
```

```python
gtDF3=gtDF3.drop(gtDF3[gtDF3['기준일자']=='전체'].index)
gtDF3=gtDF3.drop(gtDF3[gtDF3['교통수단구분']=='전체'].index) 
gt3=gtDF3.groupby(['교통수단구분'])['인원'].sum()
gt3=pd.DataFrame(gt3)
gt3

plt.figure(figsize=(15,6))
plt.bar(gt3.index, gt3.loc[gt3.index, '인원'], color='#96ceb4') 
plt.title('국민 해외관광객 교통수단별')
plt.ticklabel_format(axis='y',useOffset=False, style='plain') 
```
